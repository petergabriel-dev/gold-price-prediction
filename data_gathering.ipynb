```python
import pandas as pd
from binance.client import Client
from datetime import datetime, timedelta
```

```python
client = Client()
print("✓ Binance client initialized successfully!")
```

```python
symbol = 'PAXGUSDT'
interval = '4h'
days = 180
```

```python
start_date = (datetime.now() - timedelta(days=days)).strftime("%d %b %Y")

print(f"Fetching {symbol} data...")
print(f"Interval: {interval} (daily candles)")
print(f"Period: Last {days} days")
print(f"Start date: {start_date}\n")
```

```python
klines = client.get_historical_klines(
    symbol, 
    interval, 
    start_date
)
```

```python
df = pd.DataFrame(klines, columns=[
    'timestamp', 'open', 'high', 'low', 'close', 'volume',
    'close_time', 'quote_volume', 'trades', 'taker_buy_base',
    'taker_buy_quote', 'ignore'
])
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
df ['close_time'] = pd.to_datetime(df['close_time'], unit='ms')
```

```python
df.head()
```

```python
for col in ['open', 'high', 'low', 'close', 'volume']:
    df[col] = df[col].astype(float)
```

```python
df.set_index('timestamp', inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
print("\n" + "="*60)
print("DATA PREVIEW")
print("="*60)
print(f"\nDataFrame shape: {df.shape}")
print(f"Date range: {df.index[0]} to {df.index[-1]}")
print(f"\nFirst 5 rows:")
print(df[['open', 'high', 'low', 'close', 'volume', 'close_time']].head())
print(f"\nLast 5 rows:")
print(df[['open', 'high', 'low', 'close', 'volume', 'close_time']].tail())
```

```python
print("\n" + "="*60)
print("BASIC STATISTICS")
print("="*60)
print(f"\nCurrent Gold Price (PAXG): ${df['close'].iloc[-1]:.2f}")
print(f"Average Price: ${df['close'].mean():.2f}")
print(f"Highest Price: ${df['close'].max():.2f}")
print(f"Lowest Price: ${df['close'].min():.2f}")
print(f"Price Range: ${df['close'].max() - df['close'].min():.2f}")
```

```python
df.to_csv('gold_price_data.csv')
print(f"\n✓ Data saved to 'gold_price_data.csv'")
```

```python
print("\n4. Time Continuity Check:")
date_diff = df.index.to_series().diff()
expected_gap = pd.Timedelta(days=1)
gaps = date_diff[date_diff > expected_gap * 1.5]
print(f"Number of gaps found: {len(gaps)}")
if len(gaps) == 0:
    print("✓ No significant gaps in data!")
else:
    print("⚠ Some date gaps found (may be normal for weekends):")
    print(gaps.head())
```

```python
# 5. Check for extreme outliers (price spikes > 10% in one day)
print("\n5. Extreme Price Movement Check:")
df['price_change_pct'] = df['close'].pct_change() * 100
extreme_changes = df[abs(df['price_change_pct']) > 10]
print(f"Days with >10% price change: {len(extreme_changes)}")
if len(extreme_changes) > 0:
    print("Large movements detected:")
    print(extreme_changes[['close', 'price_change_pct']].head())
else:
    print("✓ No extreme outliers detected")
```
